```python
# Dependencies
import tweepy
import numpy as np
import pandas as pd
from pprint import pprint
import matplotlib.pyplot as plt
import time
from matplotlib import style
%matplotlib inline
# Import and Initialize Sentiment Analyzer
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()

# Twitter API Keys
from config import (consumer_key, 
                    consumer_secret, 
                    access_token, 
                    access_token_secret)

# Setup Tweepy API Authentication
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth, parser=tweepy.parsers.JSONParser())
```

```python
# Target Search Term
target_terms = ("@BBC", "@CBS", "@CNN", "@Fox", "@nytimes")

# Variables for holding sentiments
sentiments = []

# List to hold unique IDs
unique_ids = []



for target in target_terms:
    
    # Create variable for holding the oldest tweet
    oldest_tweet = None
    
    # Counter
    counter = 1
        # Loop through 5 pages that is 100 tweets
    for x in range(5):
   
    # Get all tweets from home feed
        public_tweets = api.user_timeline(target, page=x, result_type="recent", max_id=oldest_tweet)

        for tweets in public_tweets:
            tweet_id = tweets["id"]
            results = analyzer.polarity_scores(tweets["text"])
            compound = results["compound"]
            pos = results["pos"]
            neu = results["neu"]
            neg = results["neg"]

            # Append tweet_id to ids list if it doesn't already exist
            # This allows checking for duplicate tweets
           #if tweet_id not in unique_ids:
            #    unique_ids.append(tweet_id)


            # Add sentiments for each tweet into a list
            sentiments.append({"user": target,
                            "Source Account":tweet_id,
                           "Date": tweets["created_at"], 
                           "Compound": compound,
                           "Positive": pos,
                           "Negative": neu,
                           "Neutral": neg,
                           "Tweets Ago": counter,
                            "Text": tweets["text"]})

            # Add to counter 
            counter += 1
        # Reassign the the oldest tweet (i.e. the max_id)
        # Subtract 1 so the previous oldest isn't included
        # in the new search
            oldest_tweet = tweet_id - 1
        

```

```python
#using pandas to create a dataset that
overall=pd.DataFrame(data=sentiments)

#wrting csv
overall.to_csv("tweet summary.csv", encoding='utf-8', index=False,header=True)

overall.head()

len(overall)
```

```python

aggregate=overall.groupby("user")
compound_avg=round(aggregate["Compound"].mean(), 4)
compound_avg[0]
avg=pd.DataFrame(compound_avg)
avg["user"]=avg.index
avg
overall.head()

BBC=overall.loc[overall["user"]=="@BBC"]
CBS=overall.loc[overall["user"]=="@CBS"]
CNN=overall.loc[overall["user"]=="@CNN"]
FOX=overall.loc[overall["user"]=="@Fox"]
NY=overall.loc[overall["user"]=="@nytimes"]

plt.scatter(BBC["Tweets Ago"],BBC["Compound"],marker="o",color="lightblue",alpha=0.7,edgecolor="black",linewidths=1,label="BBC")
plt.scatter(CBS["Tweets Ago"],CBS["Compound"],marker="o",color="green",alpha=0.7,edgecolor="black",linewidths=1,label="CBS")
plt.scatter(CNN["Tweets Ago"],CNN["Compound"],marker="o",color="red",alpha=0.7,edgecolor="black",linewidths=1,label="CNN")
plt.scatter(FOX["Tweets Ago"],FOX["Compound"],marker="o",color="blue",alpha=0.7,edgecolor="black",linewidths=1,label="FOX")
plt.scatter(NY["Tweets Ago"],NY["Compound"],marker="o",color="yellow",alpha=0.7,edgecolor="black",linewidths=1,label="New York Times")
plt.xlim(left=103,right=-1)
plt.grid()
plt.title("Sentiment Analysis of Meida Tweets (%s)" % time.strftime("%x"))
plt.ylabel("Tweet Polarity")
plt.xlabel("Tweets Ago")

plt.legend(title="Media Sources", bbox_to_anchor=(1.05, 1), loc=2)

plt.savefig("./Images/sentiment scatter.png")
```
